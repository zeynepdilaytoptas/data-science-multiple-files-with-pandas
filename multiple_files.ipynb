## Merging exercise

```python
inner_merged_df = a_df.merge(b_df, how='inner', on='Country')
inner_merged_df
```

## Olympic Sports and Medals, 1896-2014

```python
countries_df = pd.read_csv("dictionary.csv")
countries_df.head()
```

```python
summer_df = pd.read_csv("summer.csv")
#summer_df.head()
```

```python
#summer_df["Country"].unique()[:10]
```

```python
winter_df = pd.read_csv("winter.csv")
#winter_df.head()
```

## Concatenating Data

```python
summer_countries_df = pd.merge(
    summer_df,
    countries_df,
    left_on="Country",
    right_on="Code",
    how="inner"
)

summer_countries_df.drop(columns=["Country_x"], inplace=True)

summer_countries_df.rename(columns={"Country_y": "Country"}, inplace=True)
summer_countries_df.head()
```

```python
summer_countries_df["Season"] = "Summer"
```

```python
winter_countries_df = pd.merge(
    winter_df,
    countries_df,
    left_on="Country",
    right_on="Code",
    how="inner"
)

winter_countries_df.drop(columns=["Country_x"], inplace=True)

winter_countries_df.rename(columns={"Country_y": "Country"}, inplace=True)
winter_countries_df.head()
```

```python
winter_countries_df["Season"] = "Winter"
```

```python
all_df = pd.concat([summer_countries_df, winter_countries_df], axis="index", sort=False)
all_df
```

## Testing

```python
from nbresult import ChallengeResult

result = ChallengeResult('all_df',
    all_df_shape=all_df.shape,
    all_df_columns=set(all_df.columns)
)
result.write()

print(result.check())
```

## Analysis of 10 Most Successful Countries

```python
filtered_df = all_df[all_df["Year"] >= 1984]
grouped = filtered_df.groupby("Country").size()
top_10 = grouped.sort_values(ascending=False).head(10)
top_10_df = top_10.to_frame(name="Medal Count")
top_10_df.head()
```

```python
top_10_df.plot(kind="bar",color="lightgreen");
```

## Testing

```python
from nbresult import ChallengeResult

result = ChallengeResult('olympic_games',
    top_country_1=top_10_df.iloc[0]['Medal Count'],
    top_country_10=top_10_df.iloc[9]['Medal Count']
)
result.write()

print(result.check())
```

## Analysis of 10 Most Successful Countries (Season based)

```python
top_10_season_df = (
    all_df[all_df["Year"] >= 1984]             
    .groupby(["Country", "Season"])             
    .size()                                     
    .unstack(fill_value=0)                      
)


top_10_season_df["Total"] = (
    top_10_season_df["Summer"] + 
    top_10_season_df["Winter"]
)


top_10_season_df = (
    top_10_season_df
    .sort_values("Total", ascending=False)
    .head(10)                                   
)
```

```python
top_10_season_df[['Summer', 'Winter']].plot(kind="bar",color="magenta");
```

## Testing

```python
from nbresult import ChallengeResult

result = ChallengeResult('olympic_games_season',
    top_country_season_shape=top_10_season_df.shape,
    top_country_1_summer=top_10_season_df.iloc[0]['Summer'],
    top_country_10_winter=top_10_season_df.iloc[9]['Winter']
)
result.write()

print(result.check())
```
